```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# get the dataset
dataset = pd.read_csv("winequality-red.csv")
```

```python
dataset
```

```python
dataset.mean()
```

```python
# histogram of dependent variable
n, bins, patches = plt.hist(dataset['quality'], bins=6)
plt.show()
```

```python
# transform the dependent variable
dataset["quality"] = np.where(dataset["quality"] > 6, 1, 0)
```

```python
dataset
```

```python
# isolate X and Y variables
Y = dataset.iloc[:, -1]
X = dataset.iloc[:, :-1]
```

```python
# Logistic Regression
X = sm.add_constant(X)
X.head()
```

```python
logistic_regression = sm.Logit(Y, X).fit()
logistic_regression.summary()
```
