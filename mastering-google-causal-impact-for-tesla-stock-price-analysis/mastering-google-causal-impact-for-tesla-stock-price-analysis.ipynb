```python
import yfinance as yf
import pandas as pd
```

```python
# define the dates
training_start = "2018-01-02"
training_end = "2018-09-05"
treatment_start = "2018-09-06"
treatment_end = "2018-09-07"
end_stock = "2018-09-08"
```

```python
# get the stock data
stocks = ["TSLA", "ZAL.DE", "DIS", "BMW.DE", "GM", "KHC"]
dataset = yf.download(stocks, start=training_start, end=end_stock, interval="1d")
```

```python
dataset
```

```python
# getting only the data we want
dataset = dataset.iloc[:, :6]
dataset
```

```python
# dropping unneeded level
dataset.columns = dataset.columns.droplevel()
dataset
```

```python
# getting rid of NaN
dataset = dataset.dropna()
dataset
```

```python
# getting the dataset of only the training period
dataset_correlation = dataset[dataset.index <= training_end]
dataset_correlation
```

```python
# correlation matrix
dataset_correlation.corr()
```

```python
# finishing up control group selection
final_stocks = dataset[["TSLA", "ZAL.DE", "DIS", "GM", "KHC"]]
final_stocks
```

```python
# define the pre and post period
pre_period = [training_start, training_end]
post_period = [treatment_start, treatment_end]
```

```python
# Doing causal impact
from causalimpact import CausalImpact
```

```python
impact = CausalImpact(data=final_stocks,
                     pre_period = pre_period,
                     post_period = post_period)
```

```python
# results
print(impact.summary())
```
